```python
import numpy as np
import scipy as st
from scipy.stats import laplace
import seaborn as sns
import matplotlib.pyplot as plt

class exampleClass:
    def __init__(self, N, mean, std):
        self.dist=laplace.rvs(loc=mean,scale=std,size=N)
    def cdf(self):
        sns.lineplot(x=self.dist, y=laplace.cdf(self.dist))
        plt.show()
    def pdf(self):
        sns.lineplot(x=self.dist, y=laplace.pdf(self.dist))
        plt.show()
```

```python
example=exampleClass(5000,0,1)
ax=sns.distplot(example.dist)
```

```python
example.cdf()
```

```python
example.pdf()
```
